```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
myimg = cv2.imread("naijatech.jpeg",-1) 
help(cv2.imread)
```

```python
plt.imshow(myimg[:-1])
```

```python
rotated = cv2.getRotationMatrix2D((myimg.shape[1]/2,myimg.shape[0]/2),20, 1) #20 degrees rotation matrix
```

```python
result = cv2.warpAffine(myimg, rotated, myimg.shape[1::-1], flags=cv2.INTER_LINEAR)
```

```python
plt.imshow(result)
```
